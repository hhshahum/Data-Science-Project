```python
import pandas as pd
import numpy as np
import seaborn as sn
from matplotlib import pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
d = open('/Users/avery/Downloads/project1/data/student/final_combined_dataset.csv')
data_lst = [line.split(',')for line in d.readlines()]
data = pd.DataFrame(data_lst, columns = data_lst[0])
pd.set_option('display.max_columns', None)
data = data.drop(index = 0)
data = data.reset_index()
data = data.apply(pd.to_numeric,errors = 'ignore')
data.set_index(["index"], inplace=True)
data.head()
```

train model

```python
porG3_class = []
for porg3 in data['G3.y\n']:
    if porg3 < 10:
        porG3_class.append('F')
    elif porg3 >= 10 and porg3 < 12:
        porG3_class.append('D')
    elif porg3 >= 12 and porg3 < 14:
        porG3_class.append('C')
    elif porg3 >= 13 and porg3 < 16:
        porG3_class.append('B')
    else:
        porG3_class.append('A')
data['porG3_class'] = porG3_class
data.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
data_to_numeric = ['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian.y','nursery','internet','famsup.y']
demographic_feature = ['school','sex','age','address','famsize','Pstatus','Medu','Fedu','Mjob','Fjob','reason','guardian.y','nursery','internet','famsup.y']
data_demo = data[demographic_feature]
for i in data_to_numeric:
    data_demo[i]= label_encoder.fit_transform(data_demo[i])
data_demo.head()
```

```python
train_x = data_demo
trainG3_y = data.values[:,-1]

x_train,x_test,y_train,y_test = train_test_split(train_x,trainG3_y,test_size=0.25,random_state=666)
```

MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import cohen_kappa_score, make_scorer
GNB = MultinomialNB()
GNB1 = GNB.fit(x_train,y_train)
predGNB1 = GNB1.predict(x_test)
accuracyGNB1 = accuracy_score(y_test,predGNB1)
print("Accuracy: ",accuracyGNB1)
print(classification_report(y_test, predGNB1))
print("Kappa Score: " +str(cohen_kappa_score(predGNB1,y_test)))
```

```python
GNB3 = GNB.fit(x_train,y_train)
predGNB3 = GNB3.predict(x_test)
accuracyGNB3 = accuracy_score(y_test,predGNB3)
print("Accuracy: ",accuracyGNB3)
print(classification_report(y_test, predGNB3))
print("Kappa Score: " +str(cohen_kappa_score(predGNB3,y_test)))
```

Gussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(x_train,y_train)
predGNB = GNB.predict(x_test)
accuracyGNB = accuracy_score(y_test,predGNB)
print("Accuracy: ",accuracyGNB)
print(classification_report(y_test, predGNB))
```
