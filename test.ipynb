```python
import pandas as pd    
import numpy as np     
import seaborn as sns  
import matplotlib.pyplot as plt 
import warnings
```

```python
d = open('./student-mat.csv','r')
data_lst = [line.strip('\n').replace('"','').split(';') for line in d.readlines()]
```

```python
data_lst
```

```python
data = pd.DataFrame(data_lst, columns = data_lst[0])
```

```python
data = data.drop(index = 0)
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
#data = data.apply(pd.to_numeric,errors = 'ignore')
data
```

sex ratio

```python
sex = sns.countplot(x='sex', data=data, order=['F', 'M'])
for p in sex.patches:
    sex.annotate('{:.2f}%'.format((p.get_height() * 100) / len(data)), (p.get_x() + 0.24, p.get_height() + 2))
plt.show()
```

```python
add = sns.countplot(x='address', data=data, order=['U', 'R'])
for p in add.patches:
    add.annotate('{:.2f}%'.format((p.get_height() * 100) / len(data)), (p.get_x() + 0.24, p.get_height() + 2))
plt.show()
```

```python
Pstatus = sns.countplot(x='Pstatus', data=data, order=['T', 'A'])
for p in Pstatus.patches:
    Pstatus.annotate('{:.2f}%'.format((p.get_height() * 100) / len(data)), (p.get_x() + 0.24, p.get_height() + 2))
plt.show()
```

```python
st = sns.countplot(x='studytime', data=data, order=['1', '2', '3', '4'])
for p in st.patches:
    st.annotate('{:.2f}%'.format((p.get_height() * 100) / len(data)), (p.get_x() + 0.24, p.get_height() + 2))
plt.show()
```

```python
fig, axarr  = plt.subplots(2,figsize=(10,10))
sns.countplot(x='studytime', data=data, order=['1', '2', '3', '4'], ax=axarr[0])
sns.countplot(x='studytime', hue='address', data=data, order=['1', '2', '3', '4'],hue_order = ['U','R'], ax=axarr[1])
plt.show()
```

```python
fig, axarr  = plt.subplots(2,figsize=(10,10))
sns.countplot(x='studytime', data=data, order=['1', '2', '3', '4'], ax=axarr[0])
sns.countplot(x='studytime', hue='G3', data=data, order=['1', '2', '3', '4'], hue_order = ['U','R'], ax=axarr[1])
plt.show()
```

```python
# Import Dataset

# df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
df = data
df = data.apply(pd.to_numeric,errors = 'ignore')
# df.head()
# Plot

plt.figure(figsize=(12,10), dpi= 80)

sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)



# Decorations

plt.title('Correlogram of grades', fontsize=22)

plt.xticks(fontsize=12)

plt.yticks(fontsize=12)

plt.show()
```
