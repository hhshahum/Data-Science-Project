```python
merge=pd.read_csv("/Users/avery/Downloads/project1/data/student/merge.csv")
```

```python
G1=merge["G1.x"]
G2=merge["G2.x"]
G3=merge["G3.x"]
```

```python
plt.plot(G1,G3)
```

```python
figure,axes=plt.subplots() #得到画板、轴
axes.boxplot([G1,G2,G3],patch_artist=True)
```

```python
merge["G1.y"]
```

```python
merged=pd.DataFrame(merge,columns=["G1.x","G2.x","G3.x"])
merged
```

```python
grades=merged.pivot_table(index="G1.x",columns="G2.x",values="G3.x",aggfunc=np.sum)
grades
```

```python
plt.figure(figsize=(10,6))
ax = sns.heatmap(grades,vmin=0,vmax=20,annot=True,fmt=".1f", linewidths=0.3,cmap="RdBu_r") 
```
